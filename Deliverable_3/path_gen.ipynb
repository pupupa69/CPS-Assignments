```python
# Import needed libraries

import matplotlib.pyplot as plt
import numpy as np

import random
```

### Random world generator

```python
world = np.random.randint(2, size=(16, 16))
print(world)

plt.imshow(world, cmap="binary_r") 
plt.xticks([]) 
plt.yticks([]) 
plt.title("Creating a random 16x16 grid")
plt.show() 
```

### Slightly smarter world generator

```python
# Parameters 

num_start_points = 3    # from 1 to 16
chance_of_path = 60     # from 0 to 100

# Create a 16x16 grid filled with zeros

world = np.zeros((16, 16), dtype=int)

# Fill in the first random cells

for point in range(num_start_points):
    start = random.randint(0,15)
    world[start, 0] = 1

# Fill next ones

for j in range(1, 16):
    for i in range(16):
        if (i > 0 and world[i-1, j-1] == 1) or (world[i, j-1] == 1) or (i < 15 and world[i+1, j-1] == 1):
            if(random.randint(0,100) > (100 - chance_of_path)):
                world[i, j] = 1

# See the world

print(world)

plt.imshow(world, cmap="binary_r") 
plt.xticks([]) 
plt.yticks([]) 
plt.title("Creating a slightly smarter 16x16 grid")
plt.show() 
```
